```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("omniart_metadata.csv")
```

```python
# make all records of century of type string
df['century'] = [str(int(x)) if isinstance(x, float) else x for x in df['century']]
# filter out non numeric values of century
df = df[df['century'].apply(lambda x: x.isnumeric())]
```

```python
# make all records of century of type int
df['century'] = df['century'].astype(int)
```

```python
df
```

```python
#df.query('1 <= century <= 1')
```

```python
def cut_num(creation_year):
    length_year = len(str(creation_year))
    cut_year = 0
    if length_year == 6: #1838.0
        cut_year = float(str(creation_year)[2:4])
    if length_year == 5: #938.0
        cut_year = float(str(creation_year)[1:3])
    if length_year == 4 or length_year == 3: #93.0 or 1.0
        cut_year = abs(creation_year)
    return cut_year
```

```python
df['creation_year_cut'] = df.apply(lambda row: cut_num(row['creation_year']), axis=1)
```

```python
labels = ["'0-'5","'5-'10","'10-'15","'15-'20","'20-'25","'25-'30","'30-'35","'35-'40","'40-'45","'45-'50",
 "'50-'55","'55-'60","'60-'65","'65-'70","'70-'75","'75-'80","'80-'85","'85-'90","'90-'95","'95-'99"]
labels
```

```python
df['period'] = pd.cut(df.creation_year_cut, bins=np.arange(0, 105, 5).tolist(), include_lowest =True, labels=labels)
```

```python
df['count']=df.groupby(['century', 'period','dominant_color'])['dominant_color'].transform('count')
```

```python
res = df.groupby(['century', 'period'])[["dominant_color", "count"]].max()
```
